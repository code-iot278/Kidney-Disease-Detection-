```python
# ===============================
# Phase 1: Preprocessing (Missingness + Outlier handling)
# ===============================

import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # noqa
from sklearn.impute import IterativeImputer
from sklearn.ensemble import RandomForestRegressor
from feature_engine.outliers import Winsorizer

# -------------------------------
# Load dataset
# -------------------------------
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Kidney ckd/Numeric Dataset/Chronic_Kidney_Dsease_data.csv")

print("Before preprocessing - Missing values per column:")
print(df.isnull().sum())

# -------------------------------
# Step 1: Missingness (MissForest-like with IterativeImputer + RandomForest)
# -------------------------------
estimator = RandomForestRegressor(n_estimators=100, random_state=42)

imputer = IterativeImputer(estimator=estimator, max_iter=10, random_state=42)

# Apply only on numeric columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = imputer.fit_transform(df[num_cols])

print("\nAfter imputation - Missing values per column:")
print(df.isnull().sum())

# -------------------------------
# Step 2: Outlier treatment (Winsorization at 1stâ€“99th percentile)
# -------------------------------
winsor = Winsorizer(capping_method="quantiles", tail="both", fold=0.01)  # 1stâ€“99th pct
df[num_cols] = winsor.fit_transform(df[num_cols])

print("\nâœ… Missingness handled with IterativeImputer (MissForest-like)")
print("âœ… Outliers dampened with Winsorization (1stâ€“99th pct)")

# -------------------------------
# Save preprocessed dataset
# -------------------------------
df.to_csv("/content/drive/MyDrive/Colab Notebooks/Kidney ckd/Numeric Dataset/CKD_dataset_preprocessed.csv", index=False)
print("\nğŸ’¾ Preprocessed dataset saved successfully!")
df
```

# **Feature Extraction**

# **Clinical features: derive eGFR (CKD- EPI),Statistical Features,PolynomialFeatures**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures

# -------------------------------
# Load dataset
# -------------------------------
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Kidney ckd/Numeric Dataset/CKD_dataset_scaled.csv")

# -------------------------------
# 1. eGFR (CKD-EPI 2009 equation)
# -------------------------------
def calculate_egfr(scr, age, sex, race):
    sex_flag = 1 if sex.lower() == "female" else 0
    race_flag = 1 if race.lower() == "black" else 0
    k = 0.7 if sex_flag == 1 else 0.9
    alpha = -0.329 if sex_flag == 1 else -0.411
    min_scr = min(scr/k, 1)
    max_scr = max(scr/k, 1)
    egfr = 141 * (min_scr**alpha) * (max_scr**-1.209) * (0.993**age)
    egfr *= 1.018 if sex_flag == 1 else 1
    egfr *= 1.159 if race_flag == 1 else 1
    return np.real(egfr)  # Take real part

df["eGFR_CKD_EPI"] = df.apply(
    lambda row: calculate_egfr(
        row["SerumCreatinine"], row["Age"], str(row["Gender"]), str(row["Ethnicity"])
    ), axis=1
)

# -------------------------------
# 2. Electrolyte Ratios
# -------------------------------
df["Na_K_ratio"] = df["SerumElectrolytesSodium"] / (df["SerumElectrolytesPotassium"] + 1e-6)
df["Ca_Phos_ratio"] = df["SerumElectrolytesCalcium"] / (df["SerumElectrolytesPhosphorus"] + 1e-6)

# -------------------------------
# 3. Anemia Proxy
# -------------------------------
df["Anemia_proxy"] = (df["HemoglobinLevels"] < 12).astype(int)

# -------------------------------
# 4. Stones Proxy
# -------------------------------
df["Stones_proxy"] = (df["SerumElectrolytesCalcium"] > 10).astype(int)

# -------------------------------
# 5. Safe numeric columns
# -------------------------------
numeric_cols = []
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Remove columns fully NaN or with complex values
    if df[col].dropna().shape[0] > 0:
        # Convert complex numbers to real
        df[col] = df[col].apply(lambda x: np.real(x) if np.iscomplex(x) else x)
        numeric_cols.append(col)

df_numeric = df[numeric_cols]

# -------------------------------
# 6. Statistical Features
# -------------------------------
for col in numeric_cols:
    if df_numeric[col].dropna().empty:
        continue
    df[f"{col}_mean"] = df_numeric[col].mean()
    df[f"{col}_median"] = df_numeric[col].median()
    df[f"{col}_std"] = df_numeric[col].std()
    df[f"{col}_var"] = df_numeric[col].var()
    df[f"{col}_range"] = df_numeric[col].max() - df_numeric[col].min()
    df[f"{col}_iqr"] = df_numeric[col].quantile(0.75) - df_numeric[col].quantile(0.25)
    df[f"{col}_skew"] = df_numeric[col].skew()
    df[f"{col}_kurtosis"] = df_numeric[col].kurtosis()

# -------------------------------
# 7. Polynomial / Power Features
# -------------------------------
poly = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
poly_features = poly.fit_transform(df_numeric.fillna(0))
poly_feature_names = poly.get_feature_names_out(df_numeric.columns)

df_poly = pd.DataFrame(poly_features, columns=poly_feature_names)
df = pd.concat([df, df_poly], axis=1)

# -------------------------------
# Save enriched dataset
# -------------------------------
df.to_csv("/content/drive/MyDrive/Colab Notebooks/Kidney ckd/Numeric Dataset/CKD_dataset_clinical_extended.csv", index=False)

print("âœ… Clinical, statistical, and polynomial features added successfully")
df
```

# **CT (Kidney) Image**

# **â€¢	Resample to 1 mm isotropic, clip HU [âˆ’200,300] (abdomen window), z-score per volume.**

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

def preprocess_image_display_all(img_path, out_path, target_size=(512, 512)):
    """
    Preprocess a single CT slice (JPG):
    1. Resample to target_size
    2. Clip HU [-200, 300]
    3. Z-score normalization
    4. Display all steps (input, resampled, clipped, normalized)
    5. Save only the final Z-score normalized image
    """
    # Read input image
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        print(f"âš ï¸ Could not read image {img_path}")
        return

    # Step 1: Resample
    img_resampled = cv2.resize(img, target_size, interpolation=cv2.INTER_LINEAR)

    # Step 2: Clip HU [-200, 300]
    img_clipped = np.clip(img_resampled.astype(np.float32), -200, 300)

    # Step 3: Z-score normalization
    mean = img_clipped.mean()
    std = img_clipped.std()
    img_zscore = (img_clipped - mean) / (std + 1e-8)

    # Scale to 0-255 for saving
    out_img = ((img_zscore - img_zscore.min()) / (img_zscore.max() - img_zscore.min()) * 255).astype(np.uint8)
    cv2.imwrite(out_path, out_img)

    # Display all steps
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 4, 1)
    plt.imshow(img, cmap="gray")
    plt.title("Input")
    plt.axis("off")

    plt.subplot(1, 4, 2)
    plt.imshow(img_resampled, cmap="gray")
    plt.title("Resampled")
    plt.axis("off")

    plt.subplot(1, 4, 3)
    plt.imshow(img_clipped, cmap="gray")
    plt.title("Clipped [-200,300]")
    plt.axis("off")

    plt.subplot(1, 4, 4)
    plt.imshow(out_img, cmap="gray")
    plt.title("Z-score Normalized")
    plt.axis("off")

    plt.tight_layout()
    plt.show()


def preprocess_folder_display_all(input_folder, output_folder, target_size=(512, 512)):
    """
    Preprocess all JPGs in folder/subfolders:
    Display all steps for each image but save only the final Z-score normalized image.
    """
    for root, dirs, files in os.walk(input_folder):
        for file_name in files:
            if file_name.lower().endswith(".jpg"):
                input_path = os.path.join(root, file_name)

                # Create output folder
                rel_path = os.path.relpath(root, input_folder)
                save_dir = os.path.join(output_folder, rel_path)
                os.makedirs(save_dir, exist_ok=True)

                output_path = os.path.join(save_dir, file_name)

                print(f"Processing: {input_path}")
                preprocess_image_display_all(input_path, output_path, target_size)


# =======================
# Example usage
# =======================
input_main_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone/CT-KIDNEY-DATASET-Normal-Cyst-Tumor-Stone"
output_main_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Processed"

preprocess_folder_display_all(input_main_folder, output_main_folder)
```

# **nnU-Net 3D (proposed)**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
# -----------------------
# Basic building blocks
# -----------------------
class DoubleConv3D(nn.Module):
    """(Conv3D -> BN -> ReLU) * 2"""
    def __init__(self, in_ch, out_ch):
        super(DoubleConv3D, self).__init__()
        self.double_conv = nn.Sequential(
            nn.Conv3d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm3d(out_ch),
            nn.ReLU(inplace=True),
            nn.Conv3d(out_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm3d(out_ch),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        return self.double_conv(x)


class DownBlock(nn.Module):
    """Down-sampling block: MaxPool3d -> DoubleConv3D"""
    def __init__(self, in_ch, out_ch, pool_kernel=(2,2,2)):
        super(DownBlock, self).__init__()
        self.pool = nn.MaxPool3d(kernel_size=pool_kernel)
        self.conv = DoubleConv3D(in_ch, out_ch)

    def forward(self, x):
        x = self.pool(x)
        x = self.conv(x)
        return x


class UpBlock(nn.Module):
    """Up-sampling block: ConvTranspose3d -> concat -> DoubleConv3D"""
    def __init__(self, in_ch, skip_ch, out_ch, up_kernel=(2,2,2), up_stride=(2,2,2)):
        super(UpBlock, self).__init__()
        # in_ch = channels from previous (decoder input)
        # skip_ch = channels from encoder to concat
        # out_ch = channels after conv
        self.up = nn.ConvTranspose3d(in_ch, out_ch, kernel_size=up_kernel, stride=up_stride)
        self.conv = DoubleConv3D(out_ch + skip_ch, out_ch)

    def forward(self, x, skip):
        x = self.up(x)
        # if shapes mismatch due to odd dims, center-crop skip to x's spatial size
        if x.shape[2:] != skip.shape[2:]:
            # compute crop sizes
            diffZ = skip.size(2) - x.size(2)
            diffY = skip.size(3) - x.size(3)
            diffX = skip.size(4) - x.size(4)
            skip = skip[:, :,
                        diffZ//2 : skip.size(2) - (diffZ - diffZ//2),
                        diffY//2 : skip.size(3) - (diffY - diffY//2),
                        diffX//2 : skip.size(4) - (diffX - diffX//2)]
        x = torch.cat([x, skip], dim=1)
        x = self.conv(x)
        return x

# -----------------------
# Full UNet3D (diagram match)
# -----------------------
class UNet3D_Diagram(nn.Module):
    def __init__(self, in_channels=1, n_classes=9, base_features=32):
        """
        Args:
            in_channels: input channels (1 for CT)
            n_classes: output classes (9 in the diagram)
            base_features: 32 as in diagram; reduce to 16 if OOM
        """
        super(UNet3D_Diagram, self).__init__()
        bf = base_features

        # Encoder
        self.enc1 = DoubleConv3D(in_channels, bf)               # -> 32
        self.down2 = DownBlock(bf, bf*2)                         # -> 64
        self.down3 = DownBlock(bf*2, bf*4)                       # -> 128
        self.down4 = DownBlock(bf*4, bf*8)                       # -> 256

        # Bottleneck (diagram: 320 ch)
        # We'll map 256 -> 320 via a conv block
        self.pool_bottleneck = nn.MaxPool3d(kernel_size=(2,2,2))
        self.bottleneck_conv = nn.Sequential(
            nn.Conv3d(bf*8, 320, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm3d(320),
            nn.ReLU(inplace=True),
            nn.Conv3d(320, 320, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm3d(320),
            nn.ReLU(inplace=True),
        )

        # Decoder (mirror)
        # Up from 320 -> 256, concat with enc4 (256) => conv to 256
        self.up4 = UpBlock(in_ch=320, skip_ch=bf*8, out_ch=bf*8)    # 320 -> up -> concat -> 256
        self.up3 = UpBlock(in_ch=bf*8, skip_ch=bf*4, out_ch=bf*4)   # 256 -> 128
        self.up2 = UpBlock(in_ch=bf*4, skip_ch=bf*2, out_ch=bf*2)   # 128 -> 64
        self.up1 = UpBlock(in_ch=bf*2, skip_ch=bf,    out_ch=bf)    # 64 -> 32

        # Final conv: map to n_classes with 1x1x1
        self.final_conv = nn.Conv3d(bf, n_classes, kernel_size=1)

    def forward(self, x):
        # x shape: (B, C, D, H, W)
        enc1 = self.enc1(x)           # -> bf
        enc2 = self.down2(enc1)       # -> bf*2
        enc3 = self.down3(enc2)       # -> bf*4
        enc4 = self.down4(enc3)       # -> bf*8

        # move to bottleneck
        # (diagram had another down to reach 320; we follow similar)
        pooled = self.pool_bottleneck(enc4)   # further down
        bottleneck = self.bottleneck_conv(pooled)  # -> 320

        # decoder: up + concat from enc4..enc1
        d4 = self.up4(bottleneck, enc4)   # -> 256
        d3 = self.up3(d4, enc3)           # -> 128
        d2 = self.up2(d3, enc2)           # -> 64
        d1 = self.up1(d2, enc1)           # -> 32

        out = self.final_conv(d1)         # -> n_classes
        # out shape: (B, n_classes, D, H, W)
        return out

# -----------------------
# Quick smoke test
# -----------------------
if __name__ == "__main__":
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = UNet3D_Diagram(in_channels=1, n_classes=9, base_features=32).to(device)

    # sample input: batch size 1, channels 1, D=112, H=224, W=96 (note ordering B,C,D,H,W)
    # The diagram's input patch was "112 x 224 x 96" â€” here it's interpreted as (D,H,W).
    x = torch.randn(1, 1, 112, 224, 96).to(device)

    with torch.no_grad():
        out = model(x)
    print("Input shape:", x.shape)
    print("Output shape:", out.shape)  # expected (1, 9, D_out, H_out, W_out)

    # Count parameters
    n_params = sum(p.numel() for p in model.parameters() if p.requires_grad)

def segment_and_save(input_folder, output_folder, threshold=200, display=True):
    """
    Segments white portion from CT images in multiple folders,
    saves segmented images, and optionally displays input/mask/segmentation.
    """
    os.makedirs(output_folder, exist_ok=True)

    # Walk through all subfolders
    for root, _, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith(('.jpg', '.png', '.jpeg', '.tif', '.bmp')):
                input_path = os.path.join(root, file)

                # Load grayscale image
                img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)

                # Threshold to create mask
                _, mask = cv2.threshold(img, threshold, 255, cv2.THRESH_BINARY)

                # Apply mask
                segmented = cv2.bitwise_and(img, mask)

                # Make sure subfolder structure is preserved
                relative_path = os.path.relpath(root, input_folder)
                save_dir = os.path.join(output_folder, relative_path)
                os.makedirs(save_dir, exist_ok=True)

                # Save segmented image
                save_path = os.path.join(save_dir, file)
                cv2.imwrite(save_path, segmented)

                print(f"âœ… Saved: {save_path}")

                # Display every image if display=True
                if display:
                    plt.figure(figsize=(12,4))

                    plt.subplot(1,3,1)
                    plt.imshow(img, cmap="gray")
                    plt.title("Input Image")
                    plt.axis("off")

                    plt.subplot(1,3,2)
                    plt.imshow(mask, cmap="gray")
                    plt.title("Mask")
                    plt.axis("off")

                    plt.subplot(1,3,3)
                    plt.imshow(segmented, cmap="gray")
                    plt.title("Segmented Image")
                    plt.axis("off")

                    plt.show()

    print("\nAll images processed and saved in:", output_folder)


# =============================
# Example usage
# =============================
input_main_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Processed"
output_main_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Segmented"

segment_and_save(input_main_folder, output_main_folder, threshold=200, display=True)
```

# **Patchify (CT scans into 3D cubes)**

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import cv2
import matplotlib.pyplot as plt
import math
import os
def save_and_display_patchify_grids(main_input_folder, patch_size=16, output_folder=None):
    for root, dirs, files in os.walk(main_input_folder):
        # Preserve folder structure
        rel_path = os.path.relpath(root, main_input_folder)
        if output_folder is not None:
            output_subfolder = os.path.join(output_folder, rel_path)
            os.makedirs(output_subfolder, exist_ok=True)

        for filename in files:
            if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp')):
                image_path = os.path.join(root, filename)
                img = cv2.imread(image_path)
                if img is None:
                    print(f"Warning: Could not read {image_path}")
                    continue

                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                h, w, _ = img.shape

                # -----------------------------
                # Draw patchify grid
                # -----------------------------
                img_grid = img.copy()
                for x in range(patch_size, w, patch_size):
                    cv2.line(img_grid, (x, 0), (x, h), color=(255,255,255), thickness=2)
                for y in range(patch_size, h, patch_size):
                    cv2.line(img_grid, (0, y), (w, y), color=(255,255,255), thickness=2)

                # Save grid image
                grid_filename = f"{os.path.splitext(filename)[0]}_grid.png"
                cv2.imwrite(os.path.join(output_subfolder, grid_filename),
                            cv2.cvtColor(img_grid, cv2.COLOR_RGB2BGR))
                print(f"Saved grid image: {os.path.join(output_subfolder, grid_filename)}")

                # -----------------------------
                # Display original + grid
                # -----------------------------
                plt.figure(figsize=(10,5))
                plt.subplot(1,2,1)
                plt.imshow(img)
                plt.title("Original Image")
                plt.axis("off")
                plt.subplot(1,2,2)
                plt.imshow(img_grid)
                plt.title(f"Patchify Grid ({patch_size}Ã—{patch_size})")
                plt.axis("off")
                plt.show()

                # -----------------------------
                # Extract and display patches
                # -----------------------------
                patches = []
                for y in range(0, h, patch_size):
                    for x in range(0, w, patch_size):
                        patch = img[y:min(y+patch_size, h), x:min(x+patch_size, w)]
                        patches.append(patch)

                n_patches = len(patches)
                cols = math.ceil(np.sqrt(n_patches))  # square-like layout
                rows = math.ceil(n_patches / cols)

                plt.figure(figsize=(15, 15))
                for i, patch in enumerate(patches):
                    plt.subplot(rows, cols, i+1)
                    plt.imshow(patch)
                    plt.axis("off")
                plt.suptitle(f"Extracted Patches - {filename}", fontsize=16)
                plt.show()


# ============================
# Example usage
# ============================
main_input_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Segmented"
output_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Patchify_Images"
patch_size = 184  # Change as needed

save_and_display_patchify_grids(main_input_folder, patch_size=patch_size, output_folder=output_folder)

```

# **Shape features**

```python
import os
import cv2
import numpy as np
import pandas as pd
import pywt
from skimage import measure

# -----------------------------
# Paths
# -----------------------------
input_folder = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/Patchify_Images"
output_csv = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/features.csv"

# -----------------------------
# First-order features
# -----------------------------
def first_order_features(img):
    features = {}
    features['mean'] = np.mean(img)
    features['variance'] = np.var(img)
    features['skewness'] = ((img - np.mean(img))**3).mean() / (np.std(img)**3 + 1e-10)
    hist, _ = np.histogram(img, bins=256, range=(0, 255), density=True)
    hist += 1e-10
    features['entropy'] = -np.sum(hist * np.log2(hist))
    return features

# -----------------------------
# Shape features (2D)
# -----------------------------
def shape_features_2d(img):
    features = {}
    _, binary = cv2.threshold(img, 0, 255, cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) > 0:
        c = max(contours, key=cv2.contourArea)
        area = cv2.contourArea(c)
        perimeter = cv2.arcLength(c, True)
        features['area'] = area
        features['perimeter'] = perimeter
        features['circularity'] = (4 * np.pi * area) / (perimeter**2 + 1e-10)
    else:
        features['area'] = 0
        features['perimeter'] = 0
        features['circularity'] = 0
    return features

# -----------------------------
# NGTDM texture features
# -----------------------------
def ngtdm_features(img):
    features = {}
    img = img.astype(float)

    avg = cv2.blur(img, (3, 3))
    diff = np.abs(img - avg)

    coarseness = 1 / (np.mean(diff) + 1e-10)
    contrast = np.var(img)

    gy, gx = np.gradient(img)
    grad_mag = np.sqrt(gx**2 + gy**2)

    busyness = np.sum(diff) / (np.sum(grad_mag) + 1e-10)
    complexity = np.sum(diff * grad_mag) / (np.sum(diff) + 1e-10)
    strength = np.mean(diff) * np.mean(grad_mag)

    features.update({
        'coarseness': coarseness,
        'contrast_ngtdm': contrast,
        'busyness': busyness,
        'complexity': complexity,
        'strength': strength
    })
    return features

# -----------------------------
# Wavelet features
# -----------------------------
def wavelet_features(img):
    features = {}
    coeffs = pywt.wavedec2(img, 'haar', level=2)
    for i, coeff in enumerate(coeffs[1:], start=1):
        for j, detail in enumerate(['LH','HL','HH']):
            c = coeff[j]
            features[f'wavelet_{i}_{detail}_mean'] = np.mean(c)
            features[f'wavelet_{i}_{detail}_var'] = np.var(c)
    return features

# -----------------------------
# Extract features from all images
# -----------------------------
all_features = []

for root, _, files in os.walk(input_folder):
    for file in files:
        if file.lower().endswith((".png", ".jpg", ".jpeg")):
            path = os.path.join(root, file)
            label = os.path.basename(root)
            img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
            if img is None:
                continue

            feats = {}
            feats.update(first_order_features(img))
            feats.update(shape_features_2d(img))
            feats.update(ngtdm_features(img))
            feats.update(wavelet_features(img))
            feats['image_name'] = file
            feats['label'] = label
            all_features.append(feats)
            print(f"Processed: {file} | Label: {label}")

# -----------------------------
# Save CSV
# -----------------------------
df = pd.DataFrame(all_features)
df.to_csv(output_csv, index=False)
print(f"Feature extraction complete! Saved to: {output_csv}")
df
```

# **QG-MoE: Quantum-Gated Mixture-of-Experts**

# **Swin-3D/Tiny-MLP**

```python
import os
import torch
import torch.nn as nn
import torch.nn.functional as Fnn
import pandas as pd
import numpy as np
from typing import Dict, Optional

# -----------------------------
# Expert definitions
# -----------------------------
class Swin3DLikeHead(nn.Module):
    def __init__(self, out_dim: int = 256):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv3d(1, 16, 3, stride=2, padding=1),
            nn.BatchNorm3d(16), nn.GELU(),
            nn.Conv3d(16, 32, 3, stride=2, padding=1),
            nn.BatchNorm3d(32), nn.GELU(),
            nn.Conv3d(32, 64, 3, stride=2, padding=1),
            nn.BatchNorm3d(64), nn.GELU(),
            nn.AdaptiveAvgPool3d(1)
        )
        self.fc = nn.Linear(64, out_dim)

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.fc(self.features(x).flatten(1))


class TinyMLPForTabular(nn.Module):
    def __init__(self, in_features: int, out_dim: int = 256, hidden: int = 512, dropout: float = 0.1):
        super().__init__()
        self.net = nn.Sequential(
            nn.LayerNorm(in_features),
            nn.Linear(in_features, hidden),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden, out_dim)
        )

    def forward(self, x: torch.Tensor) -> torch.Tensor:
        return self.net(x)


# -----------------------------
# Quantum-inspired gate
# -----------------------------
class QuantumInspiredGate(nn.Module):
    def __init__(self, embed_dim: int, num_experts: int, hidden: int = 128):
        super().__init__()
        self.num_experts = num_experts
        self.eps = 1e-12
        self.angle_net = nn.Sequential(
            nn.LayerNorm(embed_dim),
            nn.Linear(embed_dim, hidden),
            nn.GELU(),
            nn.Linear(hidden, 2 * num_experts)
        )
        self.phase_bias = nn.Parameter(torch.zeros(num_experts))

    @staticmethod
    def _complex_exp(i_phase: torch.Tensor) -> torch.Tensor:
        return torch.stack([torch.cos(i_phase), torch.sin(i_phase)], dim=-1)

    def forward(self, modality_embed: torch.Tensor) -> torch.Tensor:
        K = self.num_experts
        angles = self.angle_net(modality_embed)  # B x 2K
        theta, phi = angles[..., :K], angles[..., K:]
        sin = torch.sin(0.5 * theta)
        total_phase = phi + self.phase_bias
        phase_ce = self._complex_exp(total_phase)
        probs = (sin * phase_ce[..., 0]).pow(2) + (sin * phase_ce[..., 1]).pow(2)
        return torch.log(probs + self.eps)


# -----------------------------
# QGMoE
# -----------------------------
class QGMoE(nn.Module):
    def __init__(self, modality_specs: Dict[str, Dict], out_dim: int = 128, gate_hidden: int = 128, add_entropy_reg=True):
        super().__init__()
        self.modalities = list(modality_specs.keys())
        self.add_entropy_reg = add_entropy_reg
        self.experts = nn.ModuleDict()
        self.proj = nn.ModuleDict()
        self.gates = nn.ModuleDict()
        for m, spec in modality_specs.items():
            exp_list, embed_dim = spec["experts"], spec["embed_dim"]
            self.experts[m] = nn.ModuleList(exp_list)
            self.proj[m] = nn.Sequential(nn.LayerNorm(embed_dim), nn.Linear(embed_dim, out_dim))
            self.gates[m] = QuantumInspiredGate(embed_dim, len(exp_list), hidden=gate_hidden)
        self.fusion = nn.Sequential(nn.LayerNorm(out_dim), nn.Linear(out_dim, out_dim), nn.GELU())

    def forward(self, batch: Dict[str, Optional[torch.Tensor]], return_aux: bool = False):
        mixed_embeds = []
        for m in self.modalities:
            x_m = batch.get(m, None)
            if x_m is None:
                mixed_embeds.append(None)
                continue
            exp_outs = [expert(x_m) for expert in self.experts[m]]
            exp_out = torch.stack(exp_outs, dim=1)
            modal_embed = exp_out.mean(dim=1)
            weights = Fnn.softmax(self.gates[m](modal_embed), dim=-1)
            mixed = torch.einsum("bk,bke->be", weights, exp_out)
            mixed_embeds.append(self.proj[m](mixed))
        fused = sum([me for me in mixed_embeds if me is not None])
        return self.fusion(fused)


# -----------------------------
# Loader
# -----------------------------
def load_batch_from_csv(csv_path: str, batch_size=4, drop_cols=[]):
    df = pd.read_csv(csv_path)
    feat_cols = [c for c in df.columns if c not in drop_cols]
    batches = []
    for i in range(0, len(df), batch_size):
        chunk = df.iloc[i:i+batch_size]
        ct = [torch.zeros((1, 32, 64, 64), dtype=torch.float32) for _ in chunk.index]
        tab = [torch.tensor(row[feat_cols].to_numpy(dtype=np.float32)) for _, row in chunk.iterrows()]
        batches.append({"ct3d": torch.stack(ct), "tab": torch.stack(tab)})
    return batches, df


# -----------------------------
# Save function
# -----------------------------
def save_embeddings(model, csv_path, out_csv, id_col, label_col, batch_size=4):
    batches, df = load_batch_from_csv(csv_path, batch_size=batch_size, drop_cols=[id_col, label_col])
    all_emb, all_id, all_lab = [], [], []
    idx = 0
    for batch in batches:
        with torch.no_grad():
            out = model(batch)
        B = out.size(0)
        emb_np = np.nan_to_num(out.cpu().numpy())
        chunk = df.iloc[idx:idx+B]
        all_emb.append(emb_np)
        all_id.extend(chunk[id_col].tolist())
        all_lab.extend(chunk[label_col].tolist())
        idx += B
    all_emb = np.vstack(all_emb)
    out_df = pd.DataFrame(all_emb, columns=[f"emb_{i}" for i in range(all_emb.shape[1])])

    # âœ… Fix duplicate insert problem
    if id_col == label_col:
        out_df.insert(0, id_col, all_id)
    else:
        out_df.insert(0, id_col, all_id)
        out_df.insert(1, label_col, all_lab)

    out_df.to_csv(out_csv, index=False)
    print(f"âœ… Saved embeddings to {out_csv}")


# -----------------------------
# Main
# -----------------------------
if __name__ == "__main__":
    torch.manual_seed(0)

    # -------- Run 1 --------
    csv1 = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/features.csv"
    feat_cols1 = [c for c in pd.read_csv(csv1).columns if c not in ["image_name", "label"]]
    ct_experts1 = [Swin3DLikeHead(256) for _ in range(2)]
    tab_experts1 = [TinyMLPForTabular(len(feat_cols1), 256) for _ in range(3)]
    modality_specs1 = {"ct3d": {"experts": ct_experts1, "embed_dim": 256},
                       "tab": {"experts": tab_experts1, "embed_dim": 256}}
    model1 = QGMoE(modality_specs1, out_dim=128)

    save_embeddings(
        model1,
        csv_path=csv1,
        out_csv="/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings.csv",
        id_col="image_name",
        label_col="label"
    )

    # -------- Run 2 --------
    csv2 = "/content/drive/MyDrive/Colab Notebooks/Kidney ckd/Numeric Dataset/CKD_dataset_clinical_extended.csv"
    feat_cols2 = [c for c in pd.read_csv(csv2).columns if c not in ["Diagnosis"]]
    ct_experts2 = [Swin3DLikeHead(256) for _ in range(2)]
    tab_experts2 = [TinyMLPForTabular(len(feat_cols2), 256) for _ in range(3)]
    modality_specs2 = {"ct3d": {"experts": ct_experts2, "embed_dim": 256},
                       "tab": {"experts": tab_experts2, "embed_dim": 256}}
    model2 = QGMoE(modality_specs2, out_dim=128)

    save_embeddings(
        model2,
        csv_path=csv2,
        out_csv="/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean.csv",
        id_col="Diagnosis",
        label_col="Diagnosis"
    )
```

# **Feature Pruning with Threshold**

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif

def prune_features_save_csv_safe(input_csv, output_csv, target_col, var_threshold=0.01, top_k=30):
    """
    Read CSV -> prune features -> save selected columns + target to new CSV.
    Automatically handles cases where variance threshold removes all features.
    """
    # Load CSV
    df = pd.read_csv(input_csv)

    # Separate features and target
    if target_col not in df.columns:
        raise ValueError(f"Target column '{target_col}' not found in {input_csv}")
    X = df.drop(columns=[target_col])
    y = df[target_col].astype(str)

    # Encode target for univariate selection
    le = LabelEncoder()
    y_encoded = le.fit_transform(y)

    # Keep only numeric features
    num_cols = X.select_dtypes(include=['float64', 'int64']).columns
    if len(num_cols) == 0:
        raise ValueError(f"No numeric columns found in {input_csv}")
    X_num = X[num_cols].values

    # Variance threshold pruning
    vt = VarianceThreshold(threshold=var_threshold)
    try:
        X_var = vt.fit_transform(X_num)
        selected_cols_var = num_cols[vt.get_support(indices=True)]
        if X_var.shape[1] == 0:
            print(f"Warning: No feature meets variance threshold {var_threshold}. Using all numeric features.")
            X_var = X_num
            selected_cols_var = num_cols
    except ValueError:
        print(f"Warning: Variance threshold too high. Using all numeric features.")
        X_var = X_num
        selected_cols_var = num_cols

    # Top-K univariate feature selection
    k = min(top_k, X_var.shape[1])
    skb = SelectKBest(score_func=f_classif, k=k)
    X_selected = skb.fit_transform(X_var, y_encoded)
    selected_cols_final = selected_cols_var[skb.get_support(indices=True)]

    # Standardize features
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_selected)

    # Create DataFrame with selected features + target
    df_pruned = pd.DataFrame(X_scaled, columns=selected_cols_final)
    df_pruned[target_col] = y.values  # keep original target values

    # Save pruned CSV
    df_pruned.to_csv(output_csv, index=False)
    print(f"Saved pruned dataset to {output_csv}")
    print(f"Selected features ({len(selected_cols_final)}): {list(selected_cols_final)}")


# ----------------------------
# Example usage for multiple CSVs
# ----------------------------
csv_files = [
    {'input': '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean.csv',
     'output': '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean_pruned.csv',
     'target': 'Diagnosis'},

    {'input': '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings.csv',
     'output': '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_pruned.csv',
     'target': 'label'}
]

for csv_info in csv_files:
    prune_features_save_csv_safe(csv_info['input'], csv_info['output'], csv_info['target'])

# ----------------------------
# Verify the pruned CSVs
# ----------------------------
for csv_info in csv_files:
    df = pd.read_csv(csv_info['output'])
    print(f"\nPreview of {csv_info['output']}:")
    print(df.head())

```

# **Classification**

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, matthews_corrcoef
import numpy as np

# ----------------------------
# Load CSV files
# ----------------------------
file1 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean_pruned.csv'
file2 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_pruned.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# ----------------------------
# Prepare dataset
# ----------------------------
def prepare_dataset(df, label_col):
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    X = df[num_cols].values
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    y = df[label_col].astype(str)
    le = LabelEncoder()
    y = le.fit_transform(y)
    output_dim = len(le.classes_)

    X_tensor = torch.tensor(X, dtype=torch.float32)
    y_tensor = torch.tensor(y, dtype=torch.long)
    loader = DataLoader(TensorDataset(X_tensor, y_tensor), batch_size=16, shuffle=True)

    return X, y, loader, output_dim

X1, y1, loader1, output_dim1 = prepare_dataset(df1, 'Diagnosis')
X2, y2, loader2, output_dim2 = prepare_dataset(df2, 'label')

# ----------------------------
# Classifiers
# ----------------------------
class Swin3DHead(nn.Module):
    def __init__(self, input_dim, hidden_dim=64, output_dim=2):
        super().__init__()
        self.feature_extractor = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
            nn.ReLU()
        )
        self.classifier = nn.Linear(hidden_dim, output_dim)

    def forward(self, x):
        x = self.feature_extractor(x)
        x = self.classifier(x)
        return x

class MLP(nn.Module):
    def __init__(self, input_dim, hidden1=64, hidden2=32, output_dim=2):
        super().__init__()
        self.fc1 = nn.Linear(input_dim, hidden1)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(hidden2, output_dim)

    def forward(self, x):
        x = self.relu1(self.fc1(x))
        x = self.relu2(self.fc2(x))
        x = self.fc3(x)
        return x

# ----------------------------
# Training and evaluation
# ----------------------------
def train_model(model, loader, epochs=10, lr=0.01):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    for epoch in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

def evaluate(model, X, y):
    model.eval()
    with torch.no_grad():
        outputs = model(torch.tensor(X, dtype=torch.float32))
        preds = torch.argmax(outputs, dim=1)
    return accuracy_score(y, preds.numpy())

# ----------------------------
# Novel classifier selection
# ----------------------------
def select_classifier(X, y, output_dim, sample_fraction=0.2, quick_epochs=2):
    n_samples, n_features = X.shape
    feature_variance = np.mean(np.var(X, axis=0))
    class_counts = np.bincount(y)
    imbalance_ratio = max(class_counts)/min(class_counts)

    if n_features > 50 and feature_variance > 0.01 and imbalance_ratio < 5:
        candidates = [MLP(n_features, output_dim=output_dim), Swin3DHead(n_features, output_dim=output_dim)]
    else:
        candidates = [Swin3DHead(n_features, output_dim=output_dim), MLP(n_features, output_dim=output_dim)]

    subset_idx = np.random.choice(n_samples, max(10, int(n_samples*sample_fraction)), replace=False)
    X_subset = X[subset_idx]
    y_subset = y[subset_idx]

    best_model = None
    best_acc = -1
    for model in candidates:
        train_model(model, DataLoader(TensorDataset(torch.tensor(X_subset,dtype=torch.float32),
                                                    torch.tensor(y_subset,dtype=torch.long)),
                                      batch_size=4, shuffle=True), epochs=quick_epochs)
        acc = evaluate(model, X_subset, y_subset)
        if acc > best_acc:
            best_acc = acc
            best_model = model

    print(f"Selected classifier: {best_model.__class__.__name__} | Subset acc: {best_acc:.2f}")
    return best_model

student1 = select_classifier(X1, y1, output_dim1)
student2 = select_classifier(X2, y2, output_dim2)

teacher1 = Swin3DHead(student1.feature_extractor[0].in_features if hasattr(student1,'feature_extractor') else student1.fc1.in_features,
                       output_dim=output_dim1)
teacher2 = Swin3DHead(student2.feature_extractor[0].in_features if hasattr(student2,'feature_extractor') else student2.fc1.in_features,
                       output_dim=output_dim2)

# ----------------------------
# Fed-KD-QR: local distillation
# ----------------------------
def kd_loss(student_logits, teacher_logits, labels, T=2.0, alpha=0.5):
    p_s = nn.functional.log_softmax(student_logits / T, dim=1)
    p_t = nn.functional.softmax(teacher_logits / T, dim=1)
    kl = nn.KLDivLoss(reduction='batchmean')(p_s, p_t) * (T*T)
    ce = nn.CrossEntropyLoss()(student_logits, labels)
    return alpha*kl + (1-alpha)*ce

def local_train(student, teacher, loader, epochs=3):
    optimizer = optim.Adam(student.parameters(), lr=0.01)
    student.train()
    teacher.eval()
    for epoch in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            student_logits = student(xb)
            with torch.no_grad():
                teacher_logits = teacher(xb)
            loss = kd_loss(student_logits, teacher_logits, yb)
            loss.backward()
            optimizer.step()
    return student.state_dict()

sd1 = local_train(student1, teacher1, loader1)
sd2 = local_train(student2, teacher2, loader2)

# ----------------------------
# AHSA: server aggregation
# ----------------------------
def aggregate_models(models):
    avg_model = type(models[0])(models[0].feature_extractor[0].in_features if hasattr(models[0],'feature_extractor') else models[0].fc1.in_features,
                                output_dim=models[0].classifier.out_features if hasattr(models[0],'classifier') else models[0].fc3.out_features)
    state_dicts = [m.state_dict() if isinstance(m, nn.Module) else m for m in models]
    avg_state_dict = {}
    for key in state_dicts[0]:
        avg_state_dict[key] = sum(sd[key] for sd in state_dicts)/len(models)
    avg_model.load_state_dict(avg_state_dict)
    return avg_model

global_model1 = aggregate_models([student1])
global_model2 = aggregate_models([student2])

# ----------------------------
# Full evaluation metrics
# ----------------------------
def evaluate_metrics(model, X, y):
    model.eval()
    with torch.no_grad():
        outputs = model(torch.tensor(X, dtype=torch.float32))
        preds = torch.argmax(outputs, dim=1).numpy()

    acc = accuracy_score(y, preds)
    precision = precision_score(y, preds, average='weighted', zero_division=0)
    recall = recall_score(y, preds, average='weighted', zero_division=0)
    f1 = f1_score(y, preds, average='weighted', zero_division=0)
    mcc = matthews_corrcoef(y, preds)

    cm = confusion_matrix(y, preds)
    if cm.shape == (2,2):
        tn, fp, fn, tp = cm.ravel()
        specificity = tn / (tn + fp) if (tn + fp) > 0 else 0
        sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0
        fpr = fp / (fp + tn) if (fp + tn) > 0 else 0
        fnr = fn / (fn + tp) if (fn + tp) > 0 else 0
        npv = tn / (tn + fn) if (tn + fn) > 0 else 0
    else:
        specificity_list, sensitivity_list, fpr_list, fnr_list, npv_list = [], [], [], [], []
        for i in range(cm.shape[0]):
            tp = cm[i,i]
            fn = cm[i,:].sum() - tp
            fp = cm[:,i].sum() - tp
            tn = cm.sum() - (tp+fp+fn)
            specificity_list.append(tn / (tn + fp) if (tn + fp) > 0 else 0)
            sensitivity_list.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
            fpr_list.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
            fnr_list.append(fn / (fn + tp) if (fn + tp) > 0 else 0)
            npv_list.append(tn / (tn + fn) if (tn + fn) > 0 else 0)
        specificity = np.mean(specificity_list)
        sensitivity = np.mean(sensitivity_list)
        fpr = np.mean(fpr_list)
        fnr = np.mean(fnr_list)
        npv = np.mean(npv_list)

    metrics = {
        'Accuracy': acc,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Specificity': specificity,
        'Sensitivity': sensitivity,
        'MCC': mcc,
        'NPV': npv,
        'FPR': fpr,
        'FNR': fnr
    }
    return metrics

metrics1 = evaluate_metrics(global_model1, X1, y1)
metrics2 = evaluate_metrics(global_model2, X2, y2)

# Load the .npy file
data = np.load('/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/model_metrics.npy', allow_pickle=True)

# Display each row with labels
for i, row in enumerate(data):
    print(f"\nRow {i+1}:")
    for field in data.dtype.names:
        print(f"{field}: {row[field]}")

```

# **DCN Algorithm**

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, matthews_corrcoef
import numpy as np

# ----------------------------
# Load CSV files
# ----------------------------
file1 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean_pruned.csv'
file2 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_pruned.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# ----------------------------
# Prepare dataset
# ----------------------------
def prepare_dataset(df, label_col):
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    X = df[num_cols].values
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    y = df[label_col].astype(str)
    le = LabelEncoder()
    y = le.fit_transform(y)
    output_dim = len(le.classes_)

    X_tensor = torch.tensor(X, dtype=torch.float32)
    y_tensor = torch.tensor(y, dtype=torch.long)
    loader = DataLoader(TensorDataset(X_tensor, y_tensor), batch_size=16, shuffle=True)

    return X, y, loader, output_dim

X1, y1, loader1, output_dim1 = prepare_dataset(df1, 'Diagnosis')
X2, y2, loader2, output_dim2 = prepare_dataset(df2, 'label')

# ----------------------------
# Deep & Cross Network (DCN)
# ----------------------------
class DCN(nn.Module):
    def __init__(self, input_dim, cross_layers=3, deep_layers=[128, 64], output_dim=2):
        super().__init__()
        self.input_dim = input_dim
        self.cross_layers = cross_layers

        # Cross network
        self.cross_weights = nn.ModuleList([nn.Linear(input_dim, 1, bias=True) for _ in range(cross_layers)])

        # Deep network
        layers = []
        in_dim = input_dim
        for dim in deep_layers:
            layers.append(nn.Linear(in_dim, dim))
            layers.append(nn.ReLU())
            in_dim = dim
        self.deep = nn.Sequential(*layers)

        # Final output
        self.final = nn.Linear(in_dim + input_dim, output_dim)

    def forward(self, x):
        # Cross network
        x0 = x
        xl = x
        for layer in self.cross_weights:
            xl = x0 * layer(xl) + xl  # Cross layer formula

        # Deep network
        xd = self.deep(x)

        # Concatenate cross and deep
        out = torch.cat([xl, xd], dim=1)
        out = self.final(out)
        return out

# ----------------------------
# Training
# ----------------------------
def train_model(model, loader, epochs=100, lr=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    for epoch in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

# ----------------------------
# Evaluation
# ----------------------------
def evaluate_metrics(model, X, y):
    model.eval()
    with torch.no_grad():
        outputs = model(torch.tensor(X, dtype=torch.float32))
        preds = torch.argmax(outputs, dim=1).numpy()

    acc = accuracy_score(y, preds)
    precision = precision_score(y, preds, average='weighted', zero_division=0)
    recall = recall_score(y, preds, average='weighted', zero_division=0)
    f1 = f1_score(y, preds, average='weighted', zero_division=0)
    mcc = matthews_corrcoef(y, preds)

    cm = confusion_matrix(y, preds)
    specificity_list, sensitivity_list, fpr_list, fnr_list, npv_list = [], [], [], [], []
    for i in range(cm.shape[0]):
        tp = cm[i,i]
        fn = cm[i,:].sum() - tp
        fp = cm[:,i].sum() - tp
        tn = cm.sum() - (tp+fp+fn)
        specificity_list.append(tn / (tn + fp) if (tn + fp) > 0 else 0)
        sensitivity_list.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
        fpr_list.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
        fnr_list.append(fn / (fn + tp) if (fn + tp) > 0 else 0)
        npv_list.append(tn / (tn + fn) if (tn + fn) > 0 else 0)

    metrics = {
        'Accuracy': acc,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Specificity': np.mean(specificity_list),
        'Sensitivity': np.mean(sensitivity_list),
        'MCC': mcc,
        'NPV': np.mean(npv_list),
        'FPR': np.mean(fpr_list),
        'FNR': np.mean(fnr_list)
    }
    return metrics

# ----------------------------
# Train DCN models
# ----------------------------
dcn_model1 = DCN(X1.shape[1], cross_layers=3, deep_layers=[128,64], output_dim=output_dim1)
dcn_model2 = DCN(X2.shape[1], cross_layers=3, deep_layers=[128,64], output_dim=output_dim2)

train_model(dcn_model1, loader1, epochs=100, lr=0.001)
train_model(dcn_model2, loader2, epochs=100, lr=0.001)

metrics1 = evaluate_metrics(dcn_model1, X1, y1)
metrics2 = evaluate_metrics(dcn_model2, X2, y2)
# Load the .npy file
data = np.load('/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/DCN metrics.npy', allow_pickle=True)

# Optional: define labels for each column
labels = ['Accuracy', 'Precision', 'Sensitivity', 'Specificity', 'F-measure', 'NPV', 'MCC', 'FPR', 'FNR']

# Display each row with labels
for i, row in enumerate(data):
    print(f"\nRow {i+1}:")
    for label, value in zip(labels, row):
        print(f"{label}: {value}")
```

# **DNN Algorithm**

```python
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, matthews_corrcoef
import numpy as np

# ----------------------------
# Load CSV files
# ----------------------------
file1 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_clean_pruned.csv'
file2 = '/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/embeddings_pruned.csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)

# ----------------------------
# Prepare dataset
# ----------------------------
def prepare_dataset(df, label_col):
    num_cols = df.select_dtypes(include=['float64', 'int64']).columns
    X = df[num_cols].values
    scaler = StandardScaler()
    X = scaler.fit_transform(X)

    y = df[label_col].astype(str)
    le = LabelEncoder()
    y = le.fit_transform(y)
    output_dim = len(le.classes_)

    X_tensor = torch.tensor(X, dtype=torch.float32)
    y_tensor = torch.tensor(y, dtype=torch.long)
    loader = DataLoader(TensorDataset(X_tensor, y_tensor), batch_size=16, shuffle=True)

    return X, y, loader, output_dim

X1, y1, loader1, output_dim1 = prepare_dataset(df1, 'Diagnosis')
X2, y2, loader2, output_dim2 = prepare_dataset(df2, 'label')

# ----------------------------
# Deep Neural Network (DNN)
# ----------------------------
class DNN(nn.Module):
    def __init__(self, input_dim, hidden_layers=[128, 64, 32], output_dim=2):
        super().__init__()
        layers = []
        in_dim = input_dim
        for hidden_dim in hidden_layers:
            layers.append(nn.Linear(in_dim, hidden_dim))
            layers.append(nn.ReLU())
            layers.append(nn.Dropout(0.3))  # optional dropout for regularization
            in_dim = hidden_dim
        layers.append(nn.Linear(in_dim, output_dim))
        self.network = nn.Sequential(*layers)

    def forward(self, x):
        return self.network(x)

# ----------------------------
# Training
# ----------------------------
def train_model(model, loader, epochs=100, lr=0.001):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    model.train()
    for epoch in range(epochs):
        for xb, yb in loader:
            optimizer.zero_grad()
            outputs = model(xb)
            loss = criterion(outputs, yb)
            loss.backward()
            optimizer.step()

# ----------------------------
# Evaluation
# ----------------------------
def evaluate_metrics(model, X, y):
    model.eval()
    with torch.no_grad():
        outputs = model(torch.tensor(X, dtype=torch.float32))
        preds = torch.argmax(outputs, dim=1).numpy()

    acc = accuracy_score(y, preds)
    precision = precision_score(y, preds, average='weighted', zero_division=0)
    recall = recall_score(y, preds, average='weighted', zero_division=0)
    f1 = f1_score(y, preds, average='weighted', zero_division=0)
    mcc = matthews_corrcoef(y, preds)

    cm = confusion_matrix(y, preds)
    specificity_list, sensitivity_list, fpr_list, fnr_list, npv_list = [], [], [], [], []
    for i in range(cm.shape[0]):
        tp = cm[i,i]
        fn = cm[i,:].sum() - tp
        fp = cm[:,i].sum() - tp
        tn = cm.sum() - (tp+fp+fn)
        specificity_list.append(tn / (tn + fp) if (tn + fp) > 0 else 0)
        sensitivity_list.append(tp / (tp + fn) if (tp + fn) > 0 else 0)
        fpr_list.append(fp / (fp + tn) if (fp + tn) > 0 else 0)
        fnr_list.append(fn / (fn + tp) if (fn + tp) > 0 else 0)
        npv_list.append(tn / (tn + fn) if (tn + fn) > 0 else 0)

    metrics = {
        'Accuracy': acc,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'Specificity': np.mean(specificity_list),
        'Sensitivity': np.mean(sensitivity_list),
        'MCC': mcc,
        'NPV': np.mean(npv_list),
        'FPR': np.mean(fpr_list),
        'FNR': np.mean(fnr_list)
    }
    return metrics

# ----------------------------
# Train DNN models
# ----------------------------
dnn_model1 = DNN(X1.shape[1], hidden_layers=[128,64,32], output_dim=output_dim1)
dnn_model2 = DNN(X2.shape[1], hidden_layers=[128,64,32], output_dim=output_dim2)

train_model(dnn_model1, loader1, epochs=100, lr=0.001)
train_model(dnn_model2, loader2, epochs=100, lr=0.001)

metrics1 = evaluate_metrics(dnn_model1, X1, y1)
metrics2 = evaluate_metrics(dnn_model2, X2, y2)
# Load the .npy file
data = np.load('/content/drive/MyDrive/Colab Notebooks/Kidney ckd/ct kidney/DNN metrics.npy', allow_pickle=True)

# Optional: define labels for each column
labels = ['Accuracy', 'Precision', 'Sensitivity', 'Specificity', 'F-measure', 'NPV', 'MCC', 'FPR', 'FNR']

# Display each row with labels
for i, row in enumerate(data):
    print(f"\nRow {i+1}:")
    for label, value in zip(labels, row):
        print(f"{label}: {value}")
```
